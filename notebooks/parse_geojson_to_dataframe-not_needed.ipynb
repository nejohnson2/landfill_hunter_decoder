```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open('../landfill_geojson.geojson') as f:
    data = json.load(f)
```

```python
data_dict = {
    'id' : [],
    'area' : [],
    'fac_name' : [],
    'fac_state' : [],
    'fac_street' : [],
    'task_id' : [],
    'task_user_id' : [],
    'task_runinfo__skip' : [],
    'geometry' : []
}
for feature in data['features']:
    data_dict['id'].append(feature['properties']['id'])
    data_dict['area'].append(feature['properties']['area'])
    data_dict['fac_name'].append(feature['properties']['fac_name'])
    data_dict['fac_state'].append(feature['properties']['fac_state'])
    data_dict['fac_street'].append(feature['properties']['fac_street'])
    data_dict['task_id'].append(feature['properties']['task_id'])
    data_dict['task_user_id'].append(feature['properties']['task_user_id'])    
    data_dict['task_runinfo__skip'].append(feature['properties']['task_runinfo__skip'])
    data_dict['geometry'].append(feature['geometry'])    
```

```python
df = pd.DataFrame(data_dict)
```

```python
df.to_csv('../landfill_data.csv', index=False)
```
